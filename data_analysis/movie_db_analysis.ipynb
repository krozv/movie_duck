```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df
```

```python
group = df.groupby('original_language')
group_size = group.size()
table = group_size.reset_index(name='cnt')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.axis('off')
plt.table(cellText=table.values,
          colLabels=table.columns,
          loc='center',
         cellLoc='center',)
plt.savefig('grouped_table.png', bbox_inches='tight', pad_inches=0.5)
plt.close()
```
